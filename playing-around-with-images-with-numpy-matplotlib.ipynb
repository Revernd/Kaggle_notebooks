### Load the Image

```python
import numpy as np
import matplotlib.pyplot as plt
dog = plt.imread('/kaggle/input/image-data/dog.jpeg')
plt.imshow(dog)
```

### Image Upside down

```python
plt.imshow(dog[::-1,:,:])
```

### Image mirror reflection

```python
plt.imshow(dog[:,::-1,:])
```

### Image crop/Zoom

```python
plt.imshow(dog[20:300,80: 450, : ])
```

### Image contrasting

```python
plt.imshow(np.where(dog > 150, 255, 0))
```

### Negative image

```python
plt.imshow(255-dog)
```

### Inverse Image

```python
plt.imshow(dog[:,:,::-1])
```

### Single color image

```python
dog_2 = dog.copy()
dog_2[:,:,(0,1)] = 0
dog_3 = dog.copy()
dog_3[:,:,(0,2)] = 0
dog_4 = dog.copy()
dog_4[:,:,(1,2)] = 0

plt.imshow(dog_2)
```

```python
plt.imshow(dog_3)
```

```python
plt.imshow(dog_4)
```
