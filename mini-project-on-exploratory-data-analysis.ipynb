```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the Dataset

```python
df = pd.read_csv('/kaggle/working/Data.csv')
```

### Draw the Heatmap with correlation between Rank, Marcket Captial, Price, Circulating_Supply, Volume(24h) and Change(24h).

**Steps:**
- Use seaborn
- Create a figure with figsize(15,8)
- xlabel,ylabel,title must be present
- Plot the heatmap

```python
fig, ax = plt.subplots(figsize=(15, 8))
items = ['Rank', 'Market_Capital', 'Price', 'Circulating_Supply', 'Volume(24h)', 'Change(24h)']
# Plot the Heatmap with axes (ax)
data = df[items].corr()
sns.heatmap(data, annot=True)
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Heatmap Correlation")
plt.xticks(rotation=90)
plt.show()
```

### Draw the barplot containing Top 10 currency names with highest Market Capital

**Steps:**
- Use Seaborn
- Order the cryptocurrencies is must be in Descending order (i.e,bar corresponding to cryptocurrency with highest market value must be on top of bar plot).
- xlabel,ylabel,title must be present
- Plot the barplot

```python
fig, ax = plt.subplots(figsize=(8,8))
df_sub = df.sort_values(by=['Market_Capital'], ascending=False)
data = df_sub.iloc[:10, 1:3]
# Plot the countplot with axes (ax)
sns.barplot(x=list(data['Market_Capital']),y=list(data['Currency_Name']))
plt.xlabel("Market_Capital")
plt.ylabel("Currency_Name")
plt.title("Barplot containing Top10 Currency names with highest market capital")
plt.show()
```

### Draw the violinplot that represents the graph of Change(24h)

**Steps:**
- Use seaborn
- Change(24h) should be in the y-axis
- xlabel,ylabel,title must be present
- Plot the violinplot


```python
#Write your code here
fig, ax = plt.subplots(figsize=(8,8))
# Plot the violinplot with axes (ax)
data = df['Change(24h)']
sns.violinplot(y=data)
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Violinplot for Change(24h)")
plt.show()
```

### Draw a scatterplot that represents the relation between Price and Market Capital

**Steps:**
- use seaborn
- Consider data as 50 Cryptocurrencies, with higest Market Value
- xlabel,ylabel,title must be present
- Plot the Scatterplot

```python
#Write your code here
fig, ax = plt.subplots(figsize=(8,8))
# Plot the scatterplot with axes (ax)
df_sub = df[["Price","Market_Capital"]].sort_values(by=['Market_Capital'])
data = df_sub.iloc[-50:, :]
sns.scatterplot(x=list(data['Market_Capital']),y=list(data['Price']))
plt.xlabel("Market_Capital")
plt.ylabel("Price")
plt.title("Scatterplot between Price and Market Capital")
plt.show()
```

### Draw the barplot containing Top 10 currency names with Change(24h) in descending order

**Steps:**
- Use Seaborn
- xlabel,ylabel,title must be present
- Plot the barplot

```python
#Write your code here
fig, ax = plt.subplots(figsize=(8,8))
df_sub = df[['Currency_Name', 'Change(24h)']].sort_values(by=['Change(24h)'], ascending=False)
data = df_sub.iloc[:10]
plt.xlabel("Change(24h)")
plt.ylabel("Currency_Name")
plt.title("Barplot containing Top10 Currency names with Change(24h)")
sns.barplot(x=data['Change(24h)'], y=data['Currency_Name'])
plt.show()
```
